```python
dataframe['Name'].fillna('Unknown',inplace=True)
dataframe.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = ['Phy_marks','Che_marks','EM1_marks','PPS_marks','SME_marks']
for vars in var:
    plt.figure(figsize=(6,6))
    sns.boxplot(data=dataframe, y=vars)
    plt.show()
```

```python
# Remove outliers based on a threshold
var = ['Phy_marks','Che_marks','EM1_marks','PPS_marks','SME_marks']
outliers = dataframe[(dataframe['Che_marks']<100.0) & (dataframe['Che_marks']>0.0)]
dataframe=outliers
for vars in var:
    plt.figure(figsize=(6,6))
    sns.boxplot(data=dataframe, y=vars)
    plt.show()
```
