```python
data = pd.read_csv('employee.csv')
data
```

```python
grouping = data.groupby('Age')['Salary'].describe()
grouping
```

```python
age_ranges = [0,18,30,40,60,100]
labels_g = ['A','B','C','D','E']

data['Age_G'] = pd.cut(data['Age'],bins=age_ranges,labels=labels_g)
```

```python
data
```

```python
grouping = data.groupby('Age_G')['Salary'].describe()
print(grouping)

```
